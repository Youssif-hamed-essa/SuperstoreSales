## Data Loading and cleaning

```python
# library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading 
df = pd.read_csv('train.csv')
display(df.head())
```

```python
print(df.info())
```

```python
# Convert 'Order Date' to datetime, day first, coerce errors
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    dayfirst=True,
    errors='coerce',
    infer_datetime_format=True
)

display(df[['Order Date']].head())
print('Order Date dtype:', df['Order Date'].dtype)
print('parse failures (NaT):', df['Order Date'].isna().sum())
```

```python
df['Order_Day'] = df['Order Date'].dt.day.astype('Int64')
df['Order_Month'] = df['Order Date'].dt.month.astype('Int64')
df['Order_Year'] = df['Order Date'].dt.year.astype('Int64')

display(df[['Order Date', 'Order_Day', 'Order_Month', 'Order_Year']].head())
print('parse failures (NaT):', df['Order Date'].isna().sum())
```
